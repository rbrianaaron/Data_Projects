```python
import datetime
import pandas as pd
import pandas.io.data as web
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.pyplot import savefig
%matplotlib inline

royalty_trusts=['BPT', 'CRT', 'MARPS', 'MVO']
start = datetime.datetime(2009, 12, 29)
end = datetime.datetime(2014, 12, 29)


data= pd.io.data.get_data_yahoo(royalty_trusts, start, end)['Adj Close']

for royalty_trust in royalty_trusts:
    data[royalty_trust] = data[royalty_trust] / data[royalty_trust][0]

data['Sum'] = data.sum(axis=1)
data['Sum'] = data['Sum'] / data['Sum'][0]



oil_df = pd.read_excel('C:\Users\Brian\Documents\Data_Projects\OilRoyaltyTrusts\OilPrices.xls', sheetname=0)
oil_df['Date'] = pd.to_datetime(oil_df['Date'])
oil_df = oil_df.set_index('Date')
oil_df = oil_df[start:end]
oil_df['WTC'] = oil_df['WTC'] / oil_df['WTC'][0]
data['WTC'] = oil_df['WTC']

print 'Date of five year peak prices for the royalty trusts:', data['Sum'].idxmax()

print 'Date of five year peak price for the West Texas Crude(WTC):', data['WTC'].idxmax()

print 'Percent decline of royalty trusts from 5 year peak to current:', (data['Sum'].max()- data['Sum'][end]) / data['Sum'].max()

print 'Percent decline of WTC from 5 year peak to current:', (data['WTC'].max()- data['WTC'][end]) / data['WTC'].max()

plt.clf()
data.plot(y=['Sum', 'WTC'], figsize=(10,10), title='Normalized Five Year Royalty Trust Prices Versus West Texas Crude')

plot = data.plot(y=['Sum', 'WTC'], figsize=(10,10), title='Normalized Five Year Royalty Trust Prices Versus West Texas Crude')
fig = plot.get_figure()
fig.savefig('five_year.png')
```
